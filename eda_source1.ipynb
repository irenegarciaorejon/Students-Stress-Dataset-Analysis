# EXPLORATORY DATA ANALYSIS SOURCE 1

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df= pd.read_csv('StressLevelDataset.csv')
```

### UNIVARIATE ANALYSIS 

```python
cols= df.columns.tolist()  #crec que hem de treure stree label since is target
```

```python
plt.figure(figsize=(18,12))
for i,col in enumerate(cols,1):
    plt.subplot(5,5,i)
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```

### TARGET VARIABLE

```python
plt.figure(figsize=(6,4))
sns.countplot(x='stress_level',data=df)
plt.title(f"Stress level distribution")
plt.show()
```

### BOXPLOT FOR OUTLIERS

```python
plt.figure(figsize=(18,12))
for i,col in enumerate(cols,1):
    plt.subplot(5,5,i)
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
```

### CORRELATION HEATMAP

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.corr(),annot=True, fmt='.2f', cmap='coolwarm')
plt.title("Correlation heatmap")
plt.show()
```

```python
corr= df.corr()
corr_target = corr["stress_level"].sort_values(ascending=False)
corr_target
```

### BOXPLOTS BY TARGET

```python
for col in cols:
    if col!='stress_level':
        plt.figure(figsize=(6,4))
        sns.boxplot(x='stress_level',y=col,data=df)
        plt.title(f"{col} vs Stress level")
        plt.show()
```

### PAIRPLOT (TOP 6 VARAIBLES MORE CORRELATED)

```python
top_variables= corr_target.index[1:6]
sns.pairplot(df[top_variables.tolist()+ ["stress_level"]], hue='stress_level')  
plt.show()
```
